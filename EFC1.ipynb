```python
# INSERTION SORT

def insertionSort(arr):
  for i in range(1, len(arr)):

    chave = arr[i]

    j = i-1
    while j>=0 and arr[j] > chave :
      arr[j+1] = arr[j]
      j -= 1
    arr[j+1] = chave
  return
```

```python
# SELECTION SORT

def selectionSort(arr, tamanho):
  for ind in range(tamanho):
    minimo_ind = ind

    for j in range(ind + 1, tamanho):

      if arr[j] < arr[minimo_ind]:
        minimo_ind = j
      
    (arr[ind], arr[minimo_ind]) = (arr[minimo_ind], arr[ind])
  return
```

```python
# MERGE SORT

def merge(arr, esq, meio, dir):
  n1 = meio - esq + 1
  n2 = dir - meio

  esquerda = [0] * (n1)
  direita = [0] * (n2)

  for i in range(0, n1):
    esquerda[i] = arr[esq + i]
  
  for j in range(0, n2):
    direita[j] = arr[meio + 1 + j]
  
  i = 0
  j = 0
  k = esq

  while i < n1 and j < n2:
    if esquerda[i] <= direita[j]:
      arr[k] = esquerda[i]
      i = i + 1
    else:
      arr[k] = direita[j]
      j = j + 1
    k = k + 1

  while i < n1:
    arr[k] = esquerda[i]
    i = i + 1
    k = k + 1

  while j < n2:
    arr[k] = direita[j]
    j = j + 1
    k = k + 1

def mergeSort(arr, esq, dir):
  if esq < dir:
    meio = esq + (dir - esq) // 2

    mergeSort(arr, esq, meio)
    mergeSort(arr, meio + 1, dir)
    merge(arr, esq, meio, dir)
  
  return
```

```python
#BUBBLE SORT

def bubbleSort(arr):
  n = len(arr)

  troca = False

  for i in range(n):
    for j in range(n-1, 0, -1):

      if arr[j] < arr[j - 1]:
        troca = True
        arr[j], arr[j - 1] = arr[j - 1], arr[j]

    if not troca:
      return
  
  return
```

```python
#HEAP SORT

def maxHeap(arr, n, i):
  maior = i
  esq = 2 * i + 1
  dir = 2 * i + 2

  if esq < len(arr) and arr[esq] > arr[i]:
    maior = esq
  
  if dir < len(arr) and arr[dir] > arr[maior]:
    maior = dir

  if maior != i:
    arr[i], arr[maior] = arr[maior], arr[i]
    maxHeap(arr, n, maior)

def heapSort(arr):
  n = len(arr)
  for i in range(n//2, -1, -1):
    maxHeap(arr, n, i)
  
  for i in range(-1, 0, -1):
    arr[i], arr[0] = arr[0], arr[i]

    maxHeap(arr, i, 0)
  
  return
```

```python
#QUICK SORT

def partition(arr, menor, maior):

  pivo = arr[maior]

  i = menor -1

  for j in range(menor, maior):
    if arr[j] <= pivo:
      i = i + 1
      (arr[i], arr[j]) = (arr[j], arr[i])
  
  (arr[i + 1], arr[maior]) = (arr[maior], arr[i + 1])

  return i + 1;

def quickSort(arr, menor, maior):
  if menor < maior:
    pi = partition(arr, menor, maior)

    quickSort(arr, menor, pi - 1)

    quickSort(arr, pi +1, maior)
  
  return
```
